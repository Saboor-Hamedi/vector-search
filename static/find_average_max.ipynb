```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter 
```

```python
# Generate list of 10 random integers between 1 and 100 
random_data_list =[]
for _ in range(10):
    row = [rand.randint(1,100) for _ in range(10)]
    random_data_list.append(row)
for row in random_data_list:
    print(row)
```

```python
#  Find the max in all columns of of random_data_list
max_in_columns = [max(column) for column in zip(*random_data_list)]
print("Max in each column:", max_in_columns, end="")
```

```python
data = [
    [ 25,        50,         75,         10,         55,         80,         30,         90,         15,         45 ],
    [ 25,        50,         75,         10,         55,         80,         30,         90,         15,         45 ],
    [ 25,        50,         75,         10,         55,         80,         30,         90,         15,         45 ],
    [ 12,        52,         77,         12,         57,         82,         32,         92,         17,         47 ],
    [ 12,        52,         77,         12,         57,         82,         32,         92,         17,         47 ],
    [ 12,        52,         77,         12,         57,         82,         32,         92,         17,         47 ],
    [ 12,        52,         77,         12,         57,         82,         32,         92,         17,         47 ],
    [ 30,        60,         80,         20,         65,         85,         40,         95,         25,         50 ],
    [ 30,        60,         80,         20,         65,         85,         40,         95,         25,         50 ],
    [ 30,        60,         80,         20,         65,         85,         40,         95,         25,         50 ]
]

# You can access this data directly with this variable name.
```

```python
# Matrix 
for row in data:
    # print(' '.join([str(elem) for elem in row]))
    print(' ' .join(f"{x:3.1f} " for x in row))
```

```python
row_avg = [np.mean(row) for row in data]
print("Average of each row:", row_avg)
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(row_avg) + 1), row_avg, color='skyblue')
plt.xlabel('Row')
plt.ylabel('Average Value')
plt.title('Row Averages Bar Chart')
plt.xticks(range(1, len(row_avg) + 1))
plt.show()

```

```python
col_mins = [np.min(column) for column in zip(*data)]
print("Minimum of each column:", col_mins)

# Histogram of column minimums 
plt.figure(figsize=(10, 5))
plt.hist(col_mins, bins=10, color='lightgreen', edgecolor='black')
plt.xlabel('Minimum Value')
plt.ylabel('Frequency')
plt.title('Histogram of Column Minimums')
plt.show()
```

```python
col_max = [np.max(column) for column in zip(*data)]
print("Maximum of each column:", col_max)
```

```python
col_range = [max(column) - min(column) for column in zip(*data)]
print("Range of each column:", col_range)
```

```python
# Frequencies of each element in each column 
flat_data = [x for row in data for  x in row]
freq = Counter(flat_data)
freq_plus = {k: v for k , v in freq.items() if  v > 1}
print("Frequencies of each element in the matrix:", freq_plus)
```
